# Phase 3: Data Cleaning\nLoading and cleaning the student performance dataset.\n

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('../data/student_data.csv')
df.head()\n
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]\n
```

No missing values or we can handle them if they exist.\n

```python
# Create target variable 'passed' based on final grade G3 (>= 10)
df['passed'] = (df['G3'] >= 10).astype(int)

# Drop G1, G2, G3 to prevent data leakage for predicting pass/fail
df_clean = df.drop(columns=['G1', 'G2', 'G3'])

# Encode categorical variables
df_clean = pd.get_dummies(df_clean, drop_first=True)
df_clean.head()\n
```

```python
# Save the cleaned dataset
df_clean.to_csv('../data/cleaned_student_data.csv', index=False)
print('Cleaned dataset saved!')\n
```
